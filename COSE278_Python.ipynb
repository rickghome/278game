## Conditionals — Making Decisions

If this, then that. You read it exactly like English.


```python
def order_status_message(status):
    if status == "delivered":
        return "Your order has arrived!"
    elif status == "in_transit":
        return "Your order is on the way."
    elif status == "delayed":
        return "Your order is running late. Sorry."
    else:
        return "Status unknown. Please contact support."

print(order_status_message("delivered"))
print(order_status_message("delayed"))
print(order_status_message("lost"))
```

## Generate the FreshCart Dataset

Run this cell once. It creates 500 FreshCart orders across 4 cities, 6 drivers, and 3 statuses.


```python
import pandas as pd
import numpy as np

np.random.seed(278)
n = 500

cities  = ["San Diego", "Los Angeles", "San Francisco", "Seattle"]
drivers = ["D01", "D02", "D03", "D04", "D05", "D06"]

orders = pd.DataFrame({
    "order_id":    range(10001, 10001 + n),
    "city":        np.random.choice(cities, n),
    "driver_id":   np.random.choice(drivers, n),
    "status":      np.random.choice(["delivered", "delayed", "in_transit"], n,
                                    p=[0.65, 0.15, 0.20]),
    "total":       np.round(np.random.uniform(5, 120, n), 2),
    "delivery_fee":np.round(np.random.uniform(0, 8, n), 2),
})

# Force a handful of cases where delivery_fee exceeds order total (for Q5)
for i in [12, 47, 83, 156, 299]:
    orders.at[i, "total"]        = round(np.random.uniform(1, 4), 2)
    orders.at[i, "delivery_fee"] = round(np.random.uniform(7, 8), 2)

print(f"Dataset ready: {orders.shape[0]} orders, {orders.shape[1]} columns")
orders.head()
```

### Q1 — How many orders are in each status category?

### Q2 — Which city has the highest average order value?

### Q3 — Which driver has the most delayed orders?

### Q4 — How many orders have a total under $10? What percentage is that?

```python
low_value = orders[orders["total"] < 10]
count = len(low_value)
pct   = round(count / len(orders) * 100, 1)
print(f"{count} orders under $10 — {pct}% of all orders")
```
