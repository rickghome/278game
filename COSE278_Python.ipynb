### NumPy — Describing a Distribution

NumPy gives you fast math on lists of numbers. Here's how to describe what your order totals look like as a distribution.


```python
import numpy as np

totals = orders["total"].values   # get the order totals as a numpy array

print(f"Count:   {len(totals)}")
print(f"Mean:    ${np.mean(totals):.2f}")
print(f"Median:  ${np.median(totals):.2f}")
print(f"Std dev: ${np.std(totals):.2f}")
print(f"Min:     ${np.min(totals):.2f}")
print(f"Max:     ${np.max(totals):.2f}")

# What percentage of orders fall within one standard deviation of the mean?
within_1_std = np.sum(
    (totals >= np.mean(totals) - np.std(totals)) &
    (totals <= np.mean(totals) + np.std(totals))
)
print(f"\nOrders within 1 std dev of mean: {within_1_std} ({within_1_std/len(totals)*100:.1f}%)")
```

## Generate the FreshCart Dataset

Run this cell once. It creates 500 FreshCart orders across 4 cities, 6 drivers, and 3 statuses.


```python
import pandas as pd
import numpy as np

np.random.seed(278)
n = 500

cities  = ["San Diego", "Los Angeles", "San Francisco", "Seattle"]
drivers = ["D01", "D02", "D03", "D04", "D05", "D06"]

orders = pd.DataFrame({
    "order_id":     range(10001, 10001 + n),
    "city":         np.random.choice(cities, n),
    "driver_id":    np.random.choice(drivers, n),
    "status":       np.random.choice(["delivered", "delayed", "in_transit"], n,
                                     p=[0.65, 0.15, 0.20]),
    "total":        np.round(np.random.uniform(5, 120, n), 2),
    "delivery_fee": np.round(np.random.uniform(0, 8, n), 2),
})

# Force a handful of cases where delivery_fee exceeds order total (for Q8)
for i in [12, 47, 83, 156, 299]:
    orders.at[i, "total"]        = round(np.random.uniform(1, 4), 2)
    orders.at[i, "delivery_fee"] = round(np.random.uniform(7, 8), 2)

print(f"Dataset ready: {orders.shape[0]} orders, {orders.shape[1]} columns")
orders.head()
```
